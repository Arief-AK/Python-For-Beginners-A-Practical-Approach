In Python, a function is created by using the `def` keyword followed by the name of the function and the closing round brackets `()` and colon `:`.
```python
def function_name():
```

Everything that happens inside the function must be `indented`.
```python
def function_name():
    print("Hello")  # This happens inside the function
print("Goodbye")    # This does NOT happen inside the function
```

Functions can have `input` data. These are called `arguments`.
```python
def function_name(input_data):
    print(input_data)   # This will print whatever is inside input_data
```

Functions can return `output` data as well.
```python
def function_name(input_data):
    new_data = input_data * input_data  # Square operation
    return new_data
```

In the previous snippet, we can put all of the operation commands into one function `perform_operations`, and print the results in another function `print_answers`.

````python
################ Functions ################

def add_numbers(a, b):
    return a + b

def subtract_numbers(a, b):
    return a - b

def multiply_numbers(a, b):
    return a * b

def divide_numbers(a, b):
    return a / b

def perform_operations(first_number, second_number):
    addition = add_numbers(first_number, second_number)
    subtraction = subtract_numbers(first_number, second_number)
    multiplication = multiply_numbers(first_number, second_number)
    division = divide_numbers(first_number, second_number)

    return addition, subtraction, multiplication, division

def print_answers(addition, subtraction, multiplication, division):
    print(f"Addition: {addition}")
    print(f"Subtraction: {subtraction}")
    print(f"Multiplication: {multiplication}")
    print(f"Division: {division}")

################ Main Program ################

first_number = int(input("Enter the first number: "))
second_number = int(input("Enter the second number: "))

addition, subtraction, multiplication, division = perform_operations(first_number, second_number)
print_answers(addition, subtraction, multiplication, division)
````

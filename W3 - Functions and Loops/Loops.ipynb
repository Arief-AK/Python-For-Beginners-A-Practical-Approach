# Loops
Loops are used everywhere in applications. Just like the name, loops are used to repeat things within a certain time or condition.



In Python, there are two types of loops; For and While loop.

## For loops
For loops iterate over a given sequence. Let's look at an example.

```python
from time import sleep

prime_numbers = [1, 2, 3, 5, 7]

for prime in prime_numbers:
    print(prime)    # Print the prime number
    sleep(1)        # Wait for 1 second
```

For loops can "go-over" a sequence of numbers using the "**range**" function.

### **Python counts from zero and not one!!**
By default, python counts from zero, this is because with most computer systems it starts from zero. Let's look at an example.


```python
# Print numbers from 0 to 5
for number in range(6):
    print(number)    # Print the number
    sleep(1)        # Wait for 1 second 
```

```python
from time import sleep

def count_up():
    # Count from 1 to 10
    for counter in range(0,10):
        print(counter + 1)  # Print the number
        sleep(1)            # Wait for 1 second

count_up()
```

```python
import os
import time

# Idle and Punch frames
frames = [
    r"""
      O
     /|\
     / \
   Ryu: Ready!
    """,
    r"""
     \O/
      |
     / \
   Ryu: Waiting...
    """,
    r"""
      O__
     /|
     / \
   Ryu: Hadoken!
    """,
    r"""
     \O__
      |
     / \
   Ryu: Take that!
    """
]

def clear_screen():
    os.system('cls' if os.name == 'nt' else 'clear')

def animate_action(frames, loop_count=2, frame_delay=0.4):
    for _ in range(loop_count):
        for frame in frames:
            clear_screen()
            print(frame)
            time.sleep(frame_delay)

if __name__ == "__main__":
    while True:
        animate_action(frames[:2], loop_count=1)  # Idle animation
        animate_action(frames[2:], loop_count=1)  # Punch animation
```

### 1. Introduction

What is turtle graphics? It allows us to create drawings by controlling a "turtle" with simple commands. It helps us learn programming through visuals.

```python
# Using the library Turtle
from turtle import *

# Create a window screen and set its title
screen = Screen()
screen.title("Turtle Graphics")

# Create a turtle object
t = Turtle()
t.shape("turtle")
t.color("blue")

# Move the turtle to the starting position
t.forward(10)
t.right(90)
t.forward(10)

screen.mainloop()
```
